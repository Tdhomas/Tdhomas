```python
import random
class movies_quiz:
    def __init__(self, movie, attemptsLeft):
        self.movie= movie
        self.attemptsLeft=int(attemptsLeft)
        self.number = None
        self.guess = ''
        self.guesses= ''
        self.guess_movie=''
    
    
        
    
    def play(self):
        print('++++++++++++++++++++')
        self.guess_movie=self.guess_movie + self.movie[0]
        for i in range (1,len(self.movie)-1):
            if (i ==' '):
                self.guess_movie=self.guess_movie + '  '
            self.guess_movie=self.guess_movie + '-'
        self.guess_movie=self.guess_movie + self.movie[len(self.movie)-1]
        print('MOVIE : '+ self.guess_movie)
        print('++++++++++++++++++++')
        
        while self.attemptsLeft>0 :
            print('Alphabets used : ' + str(self.guesses))
            print('Wrong Attempts Left: {}', format(self.attemptsLeft) )
            
            self.guess=str(input())
            print('Give Input :' + self.guess)
            k=0
            t=False
            for i in (self.movie):
                if(self.guess ==i):
                    t=True
                    self.guess_movie= self.guess_movie[:k] + self.guess +self.guess_movie[k+1:]
                    k+=1
                else:
                    k+=1
            if(t):
                self.guesses= self.guesses +' ' + self.guess
            else:
                self.guesses= self.guesses + ' '+'('+ self.guess+')'
                self.attemptsLeft-=1
            print('++++++++++++++++++++')
            print('MOVIE : '+ self.guess_movie)
            print('++++++++++++++++++++')
            if(self.guess_movie==self.movie):
                break
        if(self.guess_movie==self.movie):
            print( 'YOU WIN !')
        else:
            print( 'YOU LOSE !')
        print ('GAME OVER')
        print('++++++++++++++++++++')
        print('MOVIE : '+ self.movie)
        print('++++++++++++++++++++')
        
def Game():
    print ('What movie would you like to guess ?')
    movie=input().lower()
    print ('How many attempts would you like ?')
    attempts=input()
    game=movies_quiz(movie,attempts)
    game.play()
```
